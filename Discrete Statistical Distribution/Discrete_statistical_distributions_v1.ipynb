```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# Uniform

```python
from scipy.stats import randint

uniform_dist = randint(1,7) # defining the distribution

random_sample = uniform_dist.rvs(1000) # drawing a random sample of 1000

# plotting
plt.hist(random_sample, bins=6)
plt.show()
```

# Bernoulli

```python
from scipy.stats import bernoulli
p=0.1 #prob of success
bern_dist = bernoulli(p)

random_vars = bern_dist.rvs(1000) #drawing a ransom sample of 1000
random_vars=pd.Series(random_vars)

# plotting the probability
plt.bar(x=random_vars.value_counts().index.astype(str), height=random_vars.value_counts()/1000)
plt.show()
```

```python
random_vars.value_counts()/1000
```

# Binomial

```python
from scipy.stats import binom

bino_dist = binom(10, 0.1) # creates an instance of dist
events = np.linspace(0,10,11) # data points for which you need to calc prob
prob = bino_dist.pmf(events) # probabilities for such events
plt.bar(events, prob) # plotting
plt.show()
```

```python
# drawing a random sample of 1000
vals = bino_dist.rvs(1000)
plt.hist(vals)
plt.show()
```

```python
# normal approximation of binomial

bino_dist = binom(100, 0.1) # creates an instance of dist
events = np.linspace(0,100,101) # data points for which you need to calc prob
prob = bino_dist.pmf(events) # probabilities for such events
plt.bar(events, prob) # plotting
plt.xlim(0,25)
plt.show()
```

# Geometric

```python
from scipy.stats import geom
p_success = 0.1
geom_dist = geom(p_success)
events = np.linspace(0,30,31)
prob = geom_dist.pmf(events)
plt.bar(events, prob)
plt.show()
```

```python
# drawing a random sample of 1000

vals = geom_dist.rvs(10000)
plt.hist(vals, bins=100)
plt.show()
```

# Poisson

```python
from scipy.stats import poisson

lam = 6 # 6 cars per hour
pois_dist = poisson(lam)
events = np.linspace(0,36,37)
prob = pois_dist.pmf(events)

plt.bar(events, prob)
plt.xlim(left=0, right=16)
plt.show()
```

```python
# same process represented by binomial

trails=3600 # 3600 secs
p_success= 6/trails 
bino_dist = binom(trails,p_success)
events = np.linspace(0,trails,trails+1)
prob = bino_dist.pmf(events)

plt.bar(events, prob)
plt.xlim(left=0, right=16)
plt.show()
```

```python
round(pois_dist.pmf(6),3), round(bino_dist.pmf(6),3)
```

```python
# normal approximation of Poisson

lam = 10
pois_dist = poisson(lam)
events = np.linspace(0,36,37)
prob = pois_dist.pmf(events)
plt.bar(events, prob)
plt.xlim(0,25)
plt.show()
```

```python
# we can see that as we increase lambda, the closer it represents a normal distribution

lam = 100
pois_dist = poisson(lam)
events = np.linspace(0,200,201)
prob = pois_dist.pmf(events)
plt.bar(events, prob)
plt.xlim(50,150)
plt.show()
```
