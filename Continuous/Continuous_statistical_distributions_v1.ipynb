## Uniform Dist

Uniform distribution models a random variable whose outcomes are equally likely to happen. The outcomes can be discrete, like the outcomes getting form tossing a die, or continuous, like the waiting time for a bus to arrive. Thus Uniform distribution can be a discrete or continuous distribution depending on the random variable.

For discrete variables probability = 1/n

For continuous variabels: probability = 1/(b-a) where [b,a] is the range of values that the random variable can take

```python
a = 1
b=4

nums = np.random.uniform(a,b,1000) # this generates 1000 data points between 1,4 based on a uniform distribution.

# as the number of data points increases, the distributions becomes clearer
plt.hist(nums)
plt.show()
```

## Normal Distribution

```python
from scipy.stats import norm
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mean=10
std = 1
norm_dist = norm(mean,std) # mean, std dev
events = np.linspace(mean-4*std, mean+4*std, 100)
prob= norm_dist.pdf(events)
sns.lineplot(x=events, y=prob)
plt.axvline(x=mean-3*std, color="maroon", linestyle="--")
plt.axvline(x=mean-2*std, color="maroon", linestyle="--")
plt.axvline(x=mean-1*std, color="maroon", linestyle="--")
plt.axvline(x=mean+1*std, color="maroon", linestyle="--")
plt.axvline(x=mean+2*std, color="maroon", linestyle="--")
plt.axvline(x=mean+3*std, color="maroon", linestyle="--")
plt.show()
```

## Exponential

Time b/w events of a Poisson distribution. Say cars passing a gate in a given hour. Exp dist would be dist of time b/w each car passing. In a way, it is the inverse of Poisson distribution.

* Events must occur at a constant rate
* Events are independent of each other

** This is a continuous distribution as compared to Poisson which is used for discrete variables.**

```python
from scipy.stats import expon

exp_dist = expon(scale=1/6) # units hours i.e. 6 cars per hour
events=np.linspace(0,2,2000)
prob = exp_dist.pdf(events)
sns.lineplot(x=events, y=prob)
plt.axvline(x=1/6, linestyle="--", color="maroon")
plt.show()
```

```python
exp_dist.cdf(1/3)
```

## Weibull

A poisson distribution with varying lambda. Read more on this.

```python
from scipy.stats import weibull_min
import seaborn as sns
import matplotlib.pyplot as plt

weib_dist = weibull_min(scale=1/6, c=1.2) # c is kappa
events=np.linspace(0,2,2000)
prob_weibull = weib_dist.pdf(events)
prob_exponential = exp_dist.pdf(events)
sns.lineplot(x=events, y=prob_weibull, label="Weibull")
sns.lineplot(x=events, y=prob_exponential, label="Exponential")
plt.legend()
plt.show()
```

## Beta

A family of binomial distribution. Read more on this

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import beta

success = 20
faliures = 80

beta_dist = beta(a=success, b=faliures)
probs_X = np.linspace(0,1,100)
probs_1 = beta_dist.pdf(probs_X)

beta_dist = beta(a=faliures, b=success)
probs_X = np.linspace(0,1,100)
probs_2 = beta_dist.pdf(probs_X)

sns.lineplot(x=probs_X, y=probs_1, label=("a=21,b=81"))
sns.lineplot(x=probs_X, y=probs_2, label=("a=81,b=21"))
plt.legend()
plt.show()
```

```python
success = 50
faliures = 50

beta_dist = beta(a=success, b=faliures)
probs_X = np.linspace(0,1,100)
probs_1 = beta_dist.pdf(probs_X)

beta_dist = beta(a=faliures-45, b=success-45)
probs_X = np.linspace(0,1,100)
probs_2 = beta_dist.pdf(probs_X)

sns.lineplot(x=probs_X, y=probs_1, label=("a=51,b=51"))
sns.lineplot(x=probs_X, y=probs_2, label=("a=6,b=6"))
plt.legend()
plt.show()
```
